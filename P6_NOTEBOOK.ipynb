```python
def cat_levels(cat, lvl):
    """ 
    une fonction qui retourne l'élément/la catégorie classé dans l'ordre "lvl"
    """
    cat = cat.split('["')[1].split('"]')[0]
    temp = cat.split(' >> ')
    if(len(temp))<2:
        temp = [temp[0], 'None', 'None']
    return temp[lvl]

def split_1(cat):
    return cat_levels(cat,0)
def split_2(cat):
    return cat_levels(cat,1)
```

```python
# Créer deux nouvelles variables cat_1 et cat_2
data['cat_1'] = data['product_category_tree'].apply(split_1)
data['cat_2'] = data['product_category_tree'].apply(split_2)
```

## 2.1. Prétraitement   

Nous allons tester les opérations de prétraitement de texte sur un exemple de texte (contenue d'une seule cellule). Ensuite, nous allons l'appliquer sur tout le dataframe.    

Le prétraitement consiste à ces 5 étapes:

- **Suppression des chiffres**
- **Suppression des petit mots** (stopwords).
- **Tokenization :** C'est le processus de décomposition du texte donné en la plus petite unité d'une phrase appelée jeton. Les signes de ponctuation, les mots et les nombres peuvent être considérés comme des jetons. Nous pouvons vouloir trouver les fréquences des mots dans tout le texte en divisant le texte donné en jetons. Ensuite, des modèles peuvent être réalisés sur ces fréquences.  
- **Conversion des mots en minuscule**  
- **Lemmatisation :** La lemmatisation est le processus de recherche de la forme du mot apparenté dans le dictionnaire. C'est différent de Stemming (il coupe les suffixes en mots selon une certaine règle où une racine de mot n'a pas besoin d'être la même racine qu'une racine morphologique basée sur un dictionnaire, c'est juste une forme égale ou plus petite du mot. Cela révèle des incohérences concernant la souche.)
Le but de la lemmatisation est de réduire les formes flexionnelles à une forme de base commune. Contrairement à la mise en place, la lemmatisation ne coupe pas simplement les inflexions. Au lieu de cela, il utilise des bases de connaissances lexicales pour obtenir les formes de base correctes des mots.

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
def doc_clean(document, stopwords=None, keep_tags=None, # ('NN' or 'JJ')
                   lemmatizer=None):
    # 1 - supprimer les mots courtes ayant deux lettres
    shortword = re.compile(r'\b\w{1,2}\b')
    document = shortword.sub('', document)
    # 1 - Tokenisation des mots dans chaque description
    tokenizer = nltk.RegexpTokenizer(r'[A-Za-z]+')
    clean_doc = tokenizer.tokenize(document)
    # 2 - convertir en minuscule
    clean_doc = [s.lower() for s in clean_doc]
    # 3 - Garder uniquement certains tags (Nom et adjectifs par expl)
    if keep_tags is not None:
        clean_doc = [word for word,tag in nltk.pos_tag(clean_doc) if tag in keep_tags]
    if stopwords is None: stopwords=[]
    # 4 - lemmatization
    if lemmatizer is not None:
        lem_doc = lemmatizer(' '.join(clean_doc))
        clean_doc = [token.lemma_ for token in lem_doc]
    # 5 - supprimer les stopwords
    clean_doc = [s for s in clean_doc if s not in stopwords]

    return clean_doc
```

```python
from wordcloud import WordCloud

def plot_wordclouds_from_frame(frame, n_top_words, n_rows=1, figsize=(18,8),
                             backgnd_color='black', cmap='Dark2'):

    """
    Prend un groupby fait sur une série de textes (non tokenized),
(-> par exemple : word_frame = data.groupby('cat_1')['desc_clean'])
extrait les n premiers mots et trace un nuage de mots des plus fréquents
mots clés pour chaque catégorie.
    """
    fig = plt.figure(figsize=figsize)

    for i, tup in enumerate(frame,1):
        n_topic, ser_texts = tup
        # creation of a corpus of all the cleaned descriptions and product_names
        corpus = ' '.join(ser_texts.values)
        # tokenizing the words in the cleaned corpus
        tokenizer = nltk.RegexpTokenizer(r'[a-z]+')
        li_words = tokenizer.tokenize(corpus.lower())
        # counting frequency of each word
        ser_freq = pd.Series(nltk.FreqDist(li_words))

        wc = WordCloud(stopwords=None, background_color=backgnd_color,
                        colormap=cmap, max_font_size=150,
                        random_state=14)
        ser_topic = ser_freq\
            .sort_values(ascending=False)[0:n_top_words]
        wc.generate(' '.join(list(ser_topic.index)))

        n_tot = len(frame)
        n_cols = (n_tot//n_rows)+((n_tot%n_rows)>0)*1
        ax = fig.add_subplot(n_rows,n_cols,i)
        plt.imshow(wc, interpolation="bilinear")
        plt.axis("off")
        plt.tight_layout()
        plt.title(n_topic, fontweight='bold')
```

## PCA

```python
def PCA_features_reduction(X_std, var_threshold=0.9, verbose=True):
    """Return the principal components from PCA, until variance threshold."""

    from sklearn import decomposition
    
    if verbose == True:
        print("Initial number of features:", X_std.shape[1])
    
    # Processing the PCA
    pca = decomposition.PCA()
    pca.fit(X_std)
    
    # Getting the explained variance ratio for each principal component
    scree = pca.explained_variance_ratio_

    # Getting the number of principal components to reach variance thresholds
    mask = scree.cumsum() > var_threshold
    nb_selected_features = len(scree[~mask]) + 1
    if verbose == True:
        print("Number of selected features:", nb_selected_features)
    
    # Compute and displays the actual ratio of explained variance
    explained_variance_sum = scree.cumsum()[nb_selected_features-1]
    if verbose == True:
        print("Cumulative explained variance:  {:.2f}%".format(explained_variance_sum*100))
    
    # Getting the projection of the data on the first components
    X_projected = pca.transform(X_std)[:,:nb_selected_features]
    
    return X_projected
```

## 2.4 Les métriques d'évaluation numériques de la classification  

- Dans cette partie, nous allons calculer la matrice de confusion entre les true_clusters et les clusters prédits.  
- Tout d'abord, nous allons encoder true_clusters du string vers numériques pour qu'on puisse leur appliquer la matrice de confusion. Ceci en utilisant LabelEncoder().  
- Ensuite nous pouvons utiliser plusieurs métriques d'évaluations afin d'évaluer les clusters prédits. Par exemple:  
  - Précision:  la précision est la capacité du classificateur à ne pas étiqueter comme positif un échantillon qui est négatif.  
  - Recall: Le rappel est la capacité du classificateur à retrouver tous les échantillons positifs.  
  - Fscore : peut être interprété comme une moyenne harmonique pondérée de la précision et du rappel, où un score atteint sa meilleure valeur à 1 et sa pire à 0.  
  --> Nous allons utiliser tout simplement "Fscore"
  
- Nous allons aussi utiliser le module "munkres" pour obtenir la permutation des labels qui minimise les erreurs de classification. ainsi on calculera l'erreur entre classification originale et classification prédite.  
Pour information, Le module Munkres fournit une implémentation de l'algorithme Munkres (également appelé Hungarian algorithm), utile pour résoudre le problème d'affectation.

```python
def error(cluster, target_cluster, k):
    """ Compute error between cluster and target cluster
    :param cluster: predicted cluster
    :param target_cluster: target cluster
    :return: error
    """
    n = np.shape(target_cluster)[0]
    M = np.zeros((k, k))
    for i in range(k):
        for j in range(k):
            M[i][j] = np.sum(np.logical_and(cluster == i, target_cluster == j))
    m = Munkres()
    indexes = m.compute(-M)
    corresp = []
    for i in range(k):
        corresp.append(indexes[i][1])
    pred_corresp = [corresp[int(predicted)] for predicted in cluster]
    acc = np.sum(pred_corresp == target_cluster) / float(len(target_cluster))
    return acc 
```

### Fonction traitement image

```python
def traitement_image(pic):
    # chargement 
    img = np.array(Image.open(pic))
    # redimentionnement 
    img_siz = cv2.resize(img, dsize=(864, 744), interpolation=cv2.INTER_CUBIC)
    # passage en niveau de gris
    img_gray = cv2.cvtColor(img_siz,cv2.COLOR_BGR2GRAY)
    # réduction du bruit
    img_bruit = gaussian_filter(img_gray, sigma=0.5)
    # réglage du contraste
    img_egal = cv2.equalizeHist(img_bruit)
    clahe = cv2.createCLAHE(clipLimit = 5) 
    img_clahe = clahe.apply(img_egal)

    return img_clahe
```

```python
# definir la fonction qui extrait les descripteurs

def extraction_orb(liste_image):
    list_feat_orb=[]
    list_des_orb=[]
    for i in liste_image:
        img = traitement_image(i)

        #points clé (ORB)
        orb = cv2.ORB_create()
        kp, des = orb.detectAndCompute(img, mask = None) 
        list_feat_orb.append(kp)
        list_des_orb.append(des)
  
        #set des descripteurs
    frame_des=np.concatenate(list_des_orb)
    ser_descriptors = pd.Series(list_des_orb,
                                index=data.index)
    return list_des_orb, ser_descriptors , frame_des
```

```python
def get_descriptors(data):
    
    data_img = data.copy()
    list_des_orb=[]
    
    for index, row in data_img.iterrows():
    
        #prétraitement des images
        img = row['urls']
        img_trait = traitement_image(img)

        #extraction des descripteurs
        orb = cv2.ORB_create()
        kp, des = orb.detectAndCompute(img_trait, mask = None) 
        list_des_orb.append(des)

    #set des descripteurs
    des_by_img = np.asarray(list_des_orb)
    all_descriptors = np.concatenate(des_by_img, axis=0)
    
    return des_by_img , all_descriptors
```

### Création des features des images

Pour chaque image :  
- prédiction des numéros de cluster de chaque descripteur  
- création d'un histogramme = comptage pour chaque numéro de cluster du nombre de descripteurs de l'image



```python
# Creation of histograms (features)

def build_histogram(model, des, image_num):
    res = model.predict(des)
    hist = np.zeros(len(model.cluster_centers_))
    nb_des=len(des)
    if nb_des==0 : print("problème histogramme image  : ", image_num)
    for i in res:
        hist[i] += 1.0/nb_des
    return hist
```

```python
# Creation of a matrix of histograms
hist_vectors=[]

for i, image_desc in enumerate(des_by_img) : 
    hist = build_histogram(kmeans, image_desc, i) #calculates the histogram
    hist_vectors.append(hist) #histogram is the feature vector

im_features = np.asarray(hist_vectors)
```
